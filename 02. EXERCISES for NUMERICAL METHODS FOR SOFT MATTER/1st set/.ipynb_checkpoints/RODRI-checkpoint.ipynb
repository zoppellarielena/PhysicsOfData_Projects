Given the function $f(x)$
$$
f(x)= \begin{cases}
0 \quad for \quad x<T \\
1 \quad for \quad x \geq T
\end{cases}
$$
the average with respect to the PDF $\rho$ is
$$
\langle f \rangle _{\rho} = \int _{\mathbb{R_+}} {f(x) \, \rho(x) \, dx} ,
$$
where $\rho(x) = e^{-x}$, defined for $x \geq 0$.


First let's work out the analytical part. We show that the average is:
- $\langle f \rangle _{\rho} = \int _{\mathbb{R_+}} {f(x) \, \rho(x) \, dx} = \int _{T} ^{\infty} {e^{-x} \, dx} = \big[-e^{-x} \big]_{T} ^{\infty} = e^{-T} $ .

For the $\sigma^2(f)$ instead one finds:
- $\sigma ^2 (f)= \int _{\mathbb{R_+}} {[f(x) -\langle f \rangle _{\rho} ]^2 \, \rho(x) \, dx} =  \int _{T} ^{\infty} {[1-2e^{-T}]e^{-x} \, dx} + \int _{\mathbb{R_+}} {e^{-2T} \, e^{-x} \, dx}= \\ =[1-2e^{-T}] e^{-T} +e^{-2T} = e^{-T}(1-e^{-T}) = \langle f \rangle _{\rho} \, \big( \,1 - \langle f \rangle _{\rho} \, \big) \;.$

Finally, given $g(a,x)=a\,e^{-ax}$ and $F=f\rho/g$ such that $\langle f \rangle _{\rho} = \langle F \rangle _{g}$, one can find the variance of $F$: 
- $\sigma ^2 (\, a, f(x) \rho (x)\, / \, g(a,x) \,)= \int _{\mathbb{R_+}} {\big[ \, f(x)\rho (x)\, / \, g(a,x) -\langle f \rangle _{\rho} \, \big]^2 \, g(a,x) \, dx} = \\ = \int _{T} ^{\infty} {\big[ (\, \rho (x) \, / g(a,x) \,)^2 - 2\rho(x) \, \langle f \rangle_\rho \, / g(a,x) \big] g(a,x) \, dx} + \int _{\mathbb{R_+}} {\langle f \rangle_\rho ^2 \, g(a,x) \, dx}=  \int _{T} ^{\infty} {[\,e^{-2x} \, a^{-1} \,e^{ax} - 2e^{-x} \, e^{-T} \,] \, dx} + \\ + \int _{\mathbb{R_+}} {e^{-2T} \, a\,e^{-ax} \, dx}=  \big[ -(2-a)^{-1}\,a^{-1} \,e^{-x(2-a)} + 2e^{-x}\,e^{-T}  \big]_T ^\infty + \big[ -e^{-2T-ax} \big]_0 ^\infty  = \\ $ $=$<span style="font-size: 1.5em;">$ \;(\frac{e^{-T(2-a)}}{a(2-a)} )$</span> $ -  e^{-2T} $ .

 


To find $a^*$ which minimizes the variance we take the derivative:

$ \frac{1}{a^2(2-a)^2} \big[a\,(2-a) \, T \cdot e^{-T(2-a)} - e^{-T(2-a)} \cdot (2-2a) \big] = 0  \implies a\,(2-a) T = (2-2a) $


This leads to a simple second order equation:
$-a^2 T +a(2T+2)-2=0 \implies a_{1,2} = \frac{T+1  \mp \sqrt{T^2 +1 }}{T} .$
Since we are interested in the $a \in ]0,1]$ interval, we shall take the solution with the minus sign. Thus
$a^* = \frac{T+1 - \sqrt{T^2+1}}{T}$.

In the plot below $\sigma ^2 (\, a, f(x) \rho (x)\, / \, g(a,x) \,) \, = \, Var(a)$ is shown for $a \in ]0,1]$. Additionally, the minimizing $a^*$ found analytically is highlighted.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

def var(a, T):
    return (np.exp(-T * (2 - a)) / (a * (2 - a))) - np.exp(-2 * T)
def var1(a):
    return (np.exp(-T * (2 - a)) / (a * (2 - a))) - np.exp(-2 * T)
def minimizing_a(T):
    return (T+1-(T**2+1)**0.5) /T

a_values = np.linspace(1e-4, 1.0, 10**4) 

T_values = [3, 5, 10, 20]
#initial_guess = 1e-3
a_star=[]
fig, ax = plt.subplots(nrows=1, ncols=3,figsize=(20,10))
for T in T_values:
    y = var(a_values, T)
    # result = minimize(var1, initial_guess)
    # min_value = result.fun
    # min_argument = result.x
    
    min_argument = minimizing_a(T)
    min_value=var(min_argument,T)
    a_star.append(min_argument)
    ax[0].plot(a_values, y, label='T = {}'.format(T))
    ax[0].scatter(min_argument, min_value, linewidth=3)
    if T in [10,20]:
        if T == 10:
            color = "g"
        else:
            color = "r"
        ax[1].plot(a_values, y, label='T = {}'.format(T), c=color)
        ax[1].scatter(min_argument, min_value, linewidth=3, c=color)
        if T==20:
            ax[2].plot(a_values, y, label='T = {}'.format(T), c="r")
            ax[2].scatter(min_argument, min_value, linewidth=3, c="r")
    print(f"For T={T}:")
    print(f"Minimum variance: {min_value}")
    print(f"Minimizing a*: {min_argument}")
a_star=np.array(a_star)
for a in ax:
    a.set_xlabel('a')
    a.set_ylabel('Var(a)')
    a.legend()
    a.grid(True)
ax[0].set_title('Variance range for different T values')
ax[1].set_title('Variance range for T in {10, 20}')
ax[2].set_title('Variance range for T = 20')
ax[0].set_ylim(-0.001,0.012)
ax[1].set_ylim(-1e-8,1e-7)
ax[2].set_xlim(0,0.1)
ax[2].set_ylim(-1e-16,1e-15)
fig.suptitle("Variance of F and minimizing a*",fontsize=30)
plt.show()

```

It can be now noted that $\sigma^2 (f) > \sigma^2_{| a^*} (F)$ for all $ T$, as can be immediately seen in the following plot.

```python
T_range=np.linspace(1,20,100)
#a_r=(T_range+1-(T_range**2+1)**0.5) /T_range
a_r = minimizing_a(T_range)
z_F= var(a_r, T_range)
z_f=np.exp(-T_range)*(1-np.exp(-T_range))
fig, ax= plt.subplots(figsize=(10,6))
ax.plot(T_range, z_F, label="Var (F)|a*")
ax.plot(T_range, z_f, label="Var (f)")
ax.legend()
ax.grid()
# ax.set_xlim(15,20)
# ax.set_ylim(0,1e-15)
ax.set_title('Variance of f and minimal variance of F (g(a*)) compared')
ax.set_xlabel('T')
ax.set_ylabel('Var (T)')
ax.set_xticks([1,3,5,10,20])

plt.show()
```

A more thorough analysis can be performed by grouping and comparing the standard deviations. Right away $\sigma \, (\,f\,)$ and $\sigma \, (\,F\,)$ can be seen to behave quite differently: only the coefficient of variation (ratio of $\sigma \, (\,F\,)$ to $\langle F\rangle\,=\,\langle f \rangle$ ) of $F$ stays almost constanst  when $T$ varies. This behaviour is of particular interest for low averages, i.e. high $T$: the coefficient of variation of $f$ explodes for $T\geq10$. The following table summarizes the results.

```python
#s(f)/<f> = e^{-T} (1-e^{-T}) * e^T
prec=5
for t in range(4):
    sigma = var(a_star[t], T_values[t]) ** 0.5
    sigma_over_f= sigma/np.exp(-T_values[t])
    s_over_f=(np.exp(-T_values[t]) *(1-np.exp(-T_values[t])))**0.5 * np.exp(T_values[t])
    sigf_over_sigF=s_over_f*np.exp(-T_values[t])/sigma
    if t == 3:
        prec=9
    print(f"T: {T_values[t]}, a*: {np.format_float_positional(a_star[t], precision=4)}, "
          f"sigma(f)/<f>: {np.format_float_positional(s_over_f, precision=3)}, "
          f"sigma(F): {np.format_float_positional(sigma, precision=prec)}, "
          f"sigma(F)/<f>: {np.format_float_positional(sigma_over_f, precision=3)}, "
          f"sigma(f)/sigma(F): {np.format_float_positional(sigf_over_sigF, precision=3)}")
```

Finally one can look at the efficiency of sampling from $\rho$ or $g$ across multiple samples. The key advantage of using $g$ is twofold. For higher $T$ sampling from $\rho$ becomes increasingly inefficient as can be seen by the number of non zero $f(x)$ values for $T=10$ and $T=20$. As a result the estimation itself is more precise as can be seen by comparing the percent error on the integral.

```python
def load(file_path):
    with open(file_path, 'r') as file:
        text = file.read()
        lines = text.split("\n")
        data = [tuple(map(float, line.split())) for line in lines if line.strip()]
    x = [data[i][0] for i in range(len(data))]
    y = [data[i][1] for i in range(len(data))]
    return x,y
def load_globals(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    global_variables = {}
    for line in lines:
        key, value = line.strip().split(': ')
        global_variables[key] = float(value)
    return global_variables
```

```python
def f(x, T):
    return 1 if x >= T else 0

def rho(x):
    return np.exp(-x)
def g(x,a):
    return a*np.exp(-a*x)
num_samples_list = [1000,10000, 100000]
fig,ax=plt.subplots(nrows=1, ncols=3,figsize=(20,6))
for i,num_samples in enumerate (num_samples_list):
    integral_err=[]
    #plt.rcParams['text.usetex'] = False
    props = dict(boxstyle='round', facecolor='aliceblue', alpha=0.3)
    print(f"Sample size {num_samples}")
    for T in [3,5,10,20]:
        a_v = minimizing_a(T)
        samples, values=load(f"2.3_rho_T{T}_n{num_samples}.txt")
        samples=np.array(samples)
        values=np.array(values)

        integral_estimate= np.mean(values) 
        err_I=np.format_float_positional(np.abs(integral_estimate-np.exp(-T))*100/np.exp(-T) , precision=2)
        integral_err.append(err_I)
        print(f"Monte Carlo (f) estimate of the integral for T={T}:: {np.format_float_scientific(integral_estimate, precision=2)}."
        f"True value of the integral: {np.format_float_scientific(np.exp(-T), precision=2)}.")

        #ax.scatter(samples, values + np.sqrt(T)*0.02, label=f"T={T}")
        ax[i].scatter(samples[np.nonzero(values)], values[np.nonzero(values)], marker="*", label=f"T={T}, points | f(x)0: {np.count_nonzero(values)}")

    ax[i].legend(facecolor = 'aliceblue',shadow=True, edgecolor="black")
    ax[i].text(0.1, 0.16, "Sample size = {}\nIntegral estimate error [T] =\n{}%".format(num_samples, integral_err), transform=ax[i].transAxes, fontsize=15, bbox=props)
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("f(x)")
fig.suptitle("Non zero values of f(x): sampling from rho",fontsize=30)
stored_plot = (fig, ax)
#plt.show()

#plot for g
print('\nSampling from g')
fig,ax=plt.subplots(nrows=1, ncols=3,figsize=(20,6))
for i,num_samples in enumerate (num_samples_list):
    integral_err=[]
    #plt.rcParams['text.usetex'] = False
    props = dict(boxstyle='round', facecolor='aliceblue', alpha=0.3)
    print(f"Sample size {num_samples}")
    for T in [3,5,10,20]:
        a_v = minimizing_a(T)
        samples, values = load(f"2.3_g_T{T}_n{num_samples}.txt")
        samples=np.array(samples)
        values=np.array(values)
        
        integral_estimate= np.mean(values) 
        err_I=np.format_float_positional(np.abs(integral_estimate-np.exp(-T))*100/np.exp(-T) , precision=2)
        integral_err.append(err_I)
        print(f"Monte Carlo (g) estimate of the integral for T={T}: {np.format_float_scientific(integral_estimate, precision=2)}."
        f"True value of the integral: {np.format_float_scientific(np.exp(-T), precision=2)}.")

        #ax.scatter(samples, values + np.sqrt(T)*0.02, label=f"T={T}")
        ax[i].scatter(samples[np.nonzero(values)], values[np.nonzero(values)], marker="*", label=f"T={T}, points | F(x)0: {np.count_nonzero(values)}")

    ax[i].legend(facecolor = 'aliceblue',shadow=True, edgecolor="black")
    ax[i].text(0.15, 0.16, "Sample size = {}\nIntegral estimate error [T] =\n{}%".format(num_samples, integral_err), transform=ax[i].transAxes, fontsize=15, bbox=props)
    ax[i].set_xlabel("x")
    ax[i].set_ylabel("F(x)")
fig.suptitle("Non zero values of F(x): sampling from g",fontsize=30)
stored_plot1 = (fig, ax)
fig,ax=stored_plot
plt.show()
fig,ax=stored_plot1
plt.show()
```
