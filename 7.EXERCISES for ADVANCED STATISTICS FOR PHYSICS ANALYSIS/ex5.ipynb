### Uniform prior

If we have a uniform prior $g(\mu)=1 for \mu>0$, the Posterior is: 

$$
\left(\mu |\left\{y_{j}\right\}\right) \propto f\left(\left\{y_{j}\right\} \mu\right) \times g(\mu) \propto \mu^{\sum y_{j}} \mathrm{e}^{-n \mu}
$$

which corresponds to the renormalization of the Gamma function $Gamma(y|\alpha,\lambda)=k y^{\alpha -1} e^{-\lambda y}$, with $\alpha=\sum y_j+1$, $\lambda=n$ and $k=\frac{\lambda^\alpha}{\Gamma(\alpha)}$

```r
#determine and draw the posterior distribution for µ, given the data

obs <- c(4,1,3,1,3)
samples <- 1000
n <- length(obs) #n. of particles observed
range <- seq(from=0, to=8, by=1/samples) #range of μ values
alpha <- sum(obs)+1
cat("\nAlpha is:",alpha)
lambda <- n
cat("\nLambda is:",lambda)
post <- dgamma(range, alpha, n) 
plot(range, post, type="l", lwd=3, main= "Posterior distribution for Uniform Prior", xlab="μ", 
     ylab=paste("P(μ|α, n, M)"), col="maroon1")
grid()

#- evaluate mean, median and variance, both analytically and numerically in R
#analytical analysis:
mean_th <- alpha/n #mean of a gamma function
var_th <- alpha/(n^2) #variance of a gamma function
#numerical analysis:
mean_exp <- sum(range*post)/samples
var_exp <- sum((range^2)*post)/samples - (mean_exp^2)
median_exp <- qgamma(0.5, alpha, n)
err <- (abs(mean_exp-mean_th)/mean_th)*100
cat("\n\nThe analytical analysis returns\nmean:",round(mean_th,digits=6),"+/-",round(sqrt(var_th),digits=6),"\nThe median can not be calculated analytically.\n\nThe numerical analysis returns:\nmean:",round(mean_exp,digits=6),"+/-",round(sqrt(var_exp),digits=6),"\nmedian:",round(median_exp,digits=6),"\n\nThe percentual error for the numerical mean in comparison with the analytical one is:",err,"%")

# evaluate a 95% credibility interval for the results obtained with both priors. Compare the result
#with that obtained using a normal approximation for the posterior distribution, with the same mean
#and standard deviation
l1 <- qgamma(0.025, alpha, n)
l2 <- qgamma(0.975, alpha, n)
l1_norm <- qnorm(0.025, mean_th, sqrt(var_th))
l2_norm <- qnorm(0.975, mean_th, sqrt(var_th))
err1 <- (abs(l1-l1_norm)/l1)*100
err2 <- (abs(l2-l2_norm)/l2)*100
cat("\n\nFor the Uniform Prior, the 95% credibility interval is: [",l1,",",l2,"].\n\nUsing a normal approximation, instead:[",l1_norm,",",l2_norm,"].\n\nThe errors of the two approximated limits are:","\nlower:",err1,"%\nupper:",err2,"%")
lines(range,dnorm(range, mean_th, sqrt(var_th)),col="orchid",lwd=2,lty=2)
legend("topright", col=c("maroon1", "orchid"), lty=c(1,2), lwd=2, 
       legend = c("True Posterior","Gaussian Approx")
         )
```

### Jeffrey’s prior

For a Jeffrey's prior: $g(\mu) \propto \frac{1}{\sqrt{\mu}}$ for \mu>0, the Posterior is: 

$$
P\left(\mu |\left\{x_{j}\right\}\right) \propto f\left(\left\{x_{j}\right\} \mu\right) \times g(\mu) \propto \mu^{\sum x_{j}} \mathrm{e}^{-n \mu} \times \frac{1}{\sqrt{\mu}} \propto \mu^{\sum x_{j}-1 / 2} \mathrm{e}^{-n \mu}
$$

which corresponds to the renormalization of the Gamma function $Gamma(y|\alpha,\lambda)=k y^{\alpha -1} e^{-\lambda y}$, with $\alpha=\sum y_j+1/2$, $\lambda=n$ and $k=\frac{\lambda^\alpha}{\Gamma(\alpha)}$

```r
#determine and draw the posterior distribution for µ, given the data

obs <- c(4,1,3,1,3)
samples <- 100
n <- length(obs) #n. of particles observed
range <- seq(from=0, to=8, by=1/samples) #range of μ values
alpha <- sum(obs)+0.5
cat("\nAlpha is:",alpha)
lambda <- n
cat("\nLambda is:",lambda)
post <- dgamma(range, alpha, n) 
plot(range, post, type="l", lwd=3, main= "Posterior distribution for Jeffrey's Prior", xlab="μ", 
     ylab=paste("P(μ|α, n, M)"), col="green4")
grid()

#- evaluate mean, median and variance, both analytically and numerically in R
#analytical analysis:
mean_th <- alpha/n #mean of a gamma function
var_th <- alpha/(n^2) #variance of a gamma function
#numerical analysis:
mean_exp <- sum(range*post)/samples
var_exp <- sum((range^2)*post)/samples - (mean_exp^2)
median_exp <- qgamma(0.5, alpha, n)
err <- (abs(mean_exp-mean_th)/mean_th)*100
cat("\n\nThe analytical analysis returns\nmean:",round(mean_th,digits=6),"+/-",round(sqrt(var_th),digits=6),"\nThe median can not be calculated analytically.\n\nThe numerical analysis returns:\nmean:",round(mean_exp,digits=6),"+/-",round(sqrt(var_exp),digits=6),"\nmedian:",round(median_exp,digits=6),"\n\nThe percentual error for the numerical mean in comparison with the analytical one is:",err,"%")

# evaluate a 95% credibility interval for the results obtained with both priors. Compare the result
#with that obtained using a normal approximation for the posterior distribution, with the same mean
#and standard deviation
l1 <- qgamma(0.025, alpha, n)
l2 <- qgamma(0.975, alpha, n)
l1_norm <- qnorm(0.025, mean_th, sqrt(var_th))
l2_norm <- qnorm(0.975, mean_th, sqrt(var_th))
err1 <- (abs(l1-l1_norm)/l1)*100
err2 <- (abs(l2-l2_norm)/l2)*100
cat("\n\nFor the Jeffrey's Prior, the 95% credibility interval is: [",l1,",",l2,"].\n\nUsing a normal approximation, instead:[",l1_norm,",",l2_norm,"].\n\nThe errors of the two approximated limits are:","\nlower:",err1,"%\nupper:",err2,"%")
lines(range,dnorm(range, mean_th, sqrt(var_th)),col="lawngreen",lwd=2,lty=2)
legend("topright", col=c("green4", "lawngreen"), lty=c(1,2), lwd=2, 
       legend = c("True Posterior","Gaussian Approx")
         )
```

## Exercise 2


The likelihood chosen for a single signal is the uniform distribution:
$$P(\theta_k | \alpha, \beta)=\frac{1}{\pi}$$
for $\theta_k \in [-\frac{\pi}{2},\frac{\pi}{2}]$, which is the angle of emission. $\theta_k$ is also related to $\alpha$ and $\beta$ through:
$$x_{k} -\alpha = \beta \tan (\theta_{k})$$.

Changing variables on the PDF:
$$P\left(x_k | \alpha, \beta\right)=\frac{1}{\pi} \frac{\beta}{\left(x_k-\alpha \right)^{2}+\beta^{2}}$$
which is a Cauchy distribution, which is symmetric about the maximum $\alpha$ and with a FWHM of $2\beta$.
Since all the signals are independent, the total Likelihood is:

$$P\left(D | \alpha, \beta\right)=\prod_k P\left(x_k | \alpha, \beta\right)$$.
Anyway it is easier to consider the $\log{P\left(x | \alpha, \beta\right)}$:
$$\log{P\left(D | \alpha, \beta\right)}=\sum_k \log{P\left(x_k | \alpha, \beta\right)}$$

There are two unknown parameters: $\alpha$ and $\beta$, so we need two prior distributions, indipendent between each other. They are two uniform distributions:
$$
P(\alpha)=\left\{\begin{array}{cl}
\frac{1}{x_{\max }-x_{\min }} & \text { for } \alpha \in\left[x_{\min }, x_{\max }\right] \\
0 & \text { otherwise }
\end{array}\right.
$$
$$
P(\beta)=\left\{\begin{array}{cl}
\frac{1}{y_{\max }} & \text { for } \beta \in\left[0, y_{\max }\right] \\
0 & \text { otherwise }
\end{array}\right.
$$

The Posterior then is found using the Bayes' theorem: $P\left(\alpha, \beta | x_{k}\right) \propto P\left(x_{k} | \alpha, \beta\right) P(\alpha)P(\beta)$, which brings to:
$$\log \left(P\left( \alpha, \beta | D \right) \right)= \log \frac{1}{Z} \prod_k P\left(x_k | \alpha, \beta\right) = \log \frac{1}{Z} \prod_k\frac{1}{\pi} \frac{\beta}{\left(x_k-\alpha \right)^{2}+\beta^{2}} = \text{const} - \sum_k \frac{\log(\beta^2+ (x_k - \alpha))}{\beta}, $$
where Z is the normalization constant and $const$ include all the parameters that doesn't depend on $\alpha$ and $\beta$. 

The best estimation of $\alpha_o$ is given by the max of the posterior PDF.

```r
n.sample <- 1000
x.min <- -6 
x.max <- 6
h <- (x.max - x.min)/n.sample
alpha <- seq(from=x.min, by=h, length.out=n.sample)
#similarly, for beta:
y.min <- 0 
y.max <- 6
h2 <- (y.max - y.min)/n.sample
beta <- seq(from=y.min, by=h2, length.out=n.sample) #alpha and beta are two numeric types with the same length
#I will also need random chosen true values for alpha and beta
alpha.star <- runif(1, x.min, x.max)
beta.star <- runif(1, 0, y.max)
cat("The true values are:\nα:",alpha.star,"\nβ:",beta.star)
#I generate n.sample random positions x using the relation between random theta_j and x_j 
#and the true values for alpha and beta found
thetas <- runif(n.sample, min=-pi/2, max=+pi/2)
x <- beta.star*tan(thetas)+alpha.star
#Now I compute the logarithm of the posterior function, for two value a and b and a dataset x
log.posterior <- function(a, b, data) {
    return(log(dunif(a, x.min, x.max))+log(dunif(b, y.min, y.max)) + sum(log((b/(pi*(b^2 + (data - a)^2))))))
} 
#Instead of choosing a number of detections, I want to show the difference in the accuracy of the position relevation
#by increasing the number of detections
detections <- c(5,10,25,50,80,100,250,500,800,1000) 
options(repr.plot.width=15, repr.plot.height=25)  #to show two plot side by side
par(mfrow= c(5,2), mar=c(5, 5, 4, 2))
for (d in detections){
    dt = x[1:d]
    #Now I expand the posterior values to all the grid alpha x beta
    log.post_d <- outer(alpha, beta, Vectorize(function(a, b) {log.posterior(a, b, dt)}))
    post <- exp(log.post_d-max(log.post_d)) #exponential of the logarithm to find the true posterior (not renormalized)
    posterior <- post/(sum(post)*h*h2) #renormalization
    contour(alpha, beta, posterior, col ='mediumvioletred',
            xlab='α', ylab='β', main=paste("P(α,β|D) for", d,"detections")) 
    points(x = alpha.star, y = beta.star, pch = 4, col = 'royalblue3', lwd= 3) #true position 
    legend("topright", bty='n', c("True position"), col="royalblue3", pch=4)
    grid()
}
```

It can be observed how increasing the number of detections, as we would expect, the red lines, which correspond to equiprobable regions, become more and more centered to the true value.

```r
#a) vary the sampling resolution of used to generate the data, keeping the same sampling range
#  - Generative model
signal <- function (x, a, b, x0, w, t) {
    return(t * (a*exp (-(x-x0)^2/(2*w^2)) + b))
}

# Define model parameters
x0 <- 0 # Signal peak
w <- 1 # Signal width
A.true <- 2 # Signal amplitude
B.true <- 1 # Background amplitude
Delta.t <- 5 # Exposure time

# Log posterior
log.post <- function(d, x, a, b, x0, w, t) {
if(a<0 || b <0) {return(-Inf )} # the effect of the prior
sum(dpois(d, lambda=signal(x, a, b, x0, w, t), log=TRUE))
}

resolutions <- c(0.1, 0.25, 1, 2, 3)

options(repr.plot.width=15, repr.plot.height=25)  #to show the two plot side by side of each r
par(mfrow= c(5,2), mar=c(5, 5, 4, 2))

# Function to generate data 
for (r in resolutions){
    xdat <- seq(from=-7*w, to=7*w, by=r*w)
    s.true <- signal(xdat , A.true , B.true , x0, w, Delta.t)
    ddat <- rpois(length(s.true), s.true)
    xplot <- seq(from=min(xdat), to=max(xdat), by=0.05*w)
    splot <- signal(xplot , A.true , B.true , x0, w, Delta.t)
    plot(xplot , splot , xlab="x", ylab="Signal+Background counts", lwd = 2, col="purple2",
         main = paste("Signal + Background with r = ", r), type="l")
    xdat.off <- xdat - r/2
    lines(xdat.off, ddat , type="s",col="firebrick3",
    lwd=2,xlim=range(xplot), ylim=range(c(splot , ddat )))
    grid()
    
    # - Sampling grid for computing posterior
    alim <- c(0.0, A.true+2)
    blim <- c(0.5, 1.5)
    Nsamp <- 100
    uniGrid <- seq(from=1/(2*Nsamp), to=1-1/(2*Nsamp), by=1/Nsamp)
    delta_a <- diff(alim )/ Nsamp
    delta_b <- diff(blim )/ Nsamp
    a <- alim[1] + diff(alim )* uniGrid
    b <- blim[1] + diff(blim )* uniGrid
    
    # Compute log unnormalized posterior , z = ln Pˆ*(a,b|D), on a regular grid
    z <- matrix(data=NA , nrow=length(a), ncol=length(b))
    for(j in 1:length(a)) {
    for(k in 1:length(b)) {
    z[j,k] <- log.post(ddat , xdat , a[j], b[k], x0, w, Delta.t)
    }
    }
    z <- z - max(z) # set maximum to zero
    
    # Plot unnormalized 2D posterior as contours.
    contour(a, b, exp(z), nlevels = 5, labcex = 0.5, lwd = 2, xlab="amplitude , A", ylab="background , B",
           main=paste("Unnormalized 2D posterior for r=",r))
    abline(v=2,h=1,col="grey")
    grid()
}  
```

Increasing the parameter r the resolution decrease: it represent a situation in which data are few and in the unnomalized "D posterior plot it can be observed that the posterior is less centered in an amplitude value.
